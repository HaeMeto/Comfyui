```python
import os
import requests

from tqdm import tqdm

# Prompt
url = input("Masukkan URL file: ").strip()
sub_dir = input("Masukkan subfolder di dalam /workspace (kosongkan untuk root /workspace): ").strip()
sub_dir = sub_dir.lstrip("/")

# Path setup
base_dir = "/workspace"
save_dir = os.path.join(base_dir, sub_dir) if sub_dir else base_dir
os.makedirs(save_dir, exist_ok=True)

filename = url.split("/")[-1].split("?")[0]
file_path = os.path.join(save_dir, filename)

# Download
response = requests.get(url, stream=True)
total_size = int(response.headers.get('content-length', 0))

if response.status_code == 200:
    with open(file_path, 'wb') as f, tqdm(
        desc=filename,
        total=total_size,
        unit='B',
        unit_scale=True,
        unit_divisor=1024
    ) as bar:
        for chunk in response.iter_content(chunk_size=8192):
            if chunk:
                f.write(chunk)
                bar.update(len(chunk))
    print(f"\n‚úÖ File berhasil diunduh ke: {file_path}")
else:
    print(f"‚ùå Gagal mengunduh file. Status code: {response.status_code}")
```
